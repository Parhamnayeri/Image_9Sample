```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img = plt.imread("arthas.jpg")
plt.title(f"img shape {img.shape}")
plt.imshow(img)
```

```python
samples = np.zeros((9,1,300,300,3), dtype=np.uint8)
```

```python
kernel = 3
for i in range(img.shape[0]//kernel):
    for j in range(img.shape[1]//kernel):
        
        start_i = i*kernel
        end_i = (i+1)*kernel
        start_j = j*kernel
        end_j = (j+1)*kernel
        
        selectMed = img[start_i:end_i, start_j:end_j].reshape(9,1,3)
        
        for x in range(kernel**2):
            samples[x,0,i,j] = selectMed[x]
        
```

```python
fig, axes = plt.subplots(3, 3, figsize=(10, 10))
axes_flat = axes.flatten()
for i, ax in enumerate(axes_flat):
    ax.set_title(f'Sample {i+1}', fontsize=12, color='red')
    ax.imshow(samples[i,0])

plt.tight_layout()
plt.show()
```
